## Load the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras import models, layers
from keras.utils import to_categorical
```

## Import and prepare Data
Two numpy arrays (train_X, train_Y) are saved to a single .NPZ file.<br>We will read this file using the numpy.load() method and then extract the arrays.
<br><br>
As part of the holiday bonus, these datasets will require minimal processing - and we have provided the code below.<br>
Just as you did with the original MNIST exercise, you will need to perform two transformations:
1. The X dataseets will need to be reshaped from (n, 28, 28) to (n, 784), where n is the number of samples.
2. The data comes scaled to range (0, 255).  We need to rescale this to the range (0, 1).

```python
# First, read the numpy array files.
data = np.load('//ATLAFSS02/JupyterFinance/Data/ASL_MNIST/asl_mnist_train.npz')

train_X = data['train_X']
train_Y = data['train_Y']

data = 0 # clear this variable so it does not continue to consume memory.

print(train_X.shape, train_Y.shape)
```

```python
# Second, reshape the X arrays form 3D to 2D
train_X = train_X.reshape(25000, 28*28)

print(train_X.shape, train_Y.shape)
```

```python
# Third, rescale the X values to a (0, 1) range.
train_X = train_X.astype('float32') / 255.0
```

```python
# Finally, vectorize the Y values from integer to sparse 1D arrays.
print(train_Y[0])
train_Y = to_categorical(train_Y)
print(train_Y[0], train_Y[0].shape)
```
